## Pandas series 
1. It is a 1D array, holding any data type
2. It works on Single column
3. Accessing an item can be done by referring the index

```python
#indexing starts with zero
a=["this", "is", "Pandas", "Series"]
series1 = pd.Series(a)
series1
```

```python
#one can customize the labels
#"index" key word is case sensitive
a=["this", "is", "Pandas", "Series"]
series1 = pd.Series(a, index=["F1","F2","F3","F4"])
series1
```

```python
#create the series using dict
#once a label is created we can't modify it
a={"s1":[1,2,3], "s2":[3,4,5]}
dicts1=pd.Series(a)
dicts2=pd.Series(a,index=["12","21"])
print(dicts1)
print(dicts2)
```

```python
# we can select specfied keys
a={"s1":[1,2,3], "s2":[3,4,5],"s3":[6,7,8]}
dicts2=pd.Series(a,index=["s1","s3"])
print(dicts2)
```

```python
#first let us sort the data based on sqft and location
df.sort_values(by=['Area','Location'])
```

```python
#pls find the duplicates
df.duplicated().sum()
```

```python
#since there are 707 dups we are going to delete them
#inplace=true will do this operation permanently
#5014-707 = 4307
df.drop_duplicates(inplace=True)
print(df.shape)
```

```python
#since 9 in gym, swimming pool and etc is not logical we will replace those values with 1
# we can use if clause
for i in df.index:
    if df.loc[i,'SwimmingPool']==9:
        df.loc[i,'SwimmingPool']=1
        
print(df.tail(10))
```

```python
#alternate method
#df['column name'] = df['column name'].replace(['1st old value','2nd old value',...],['1st new value','2nd new value',...])
df_new = df.replace(9,1)
print(df_new)
```
