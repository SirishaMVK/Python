# Matplotlib is a data Visualization library 

```python
import matplotlib
matplotlib.__version__
```

```python
#pyplot is a subplot 
#inline will make your plot outputs appear and be stored within the notebook
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#plot a graph using numpy
import numpy as np
#exponential function graph
x=np.array([0,2,4,6,8])
y=np.array([1,4,16,36,64])
#plot function to be used
plt.plot(x,y)
#[<matplotlib.lines.Line2D at 0x1dade535cd0>] will dissappear if we use plt.show()
#plt.show()
```

```python
#plot as points
x=np.array([0,2,4,6,8])
y=np.array([1,4,16,36,64])
#plot function to be used
plt.plot(x,y,'o',color='blue')
plt.show()
```

```python
x=np.array([0,2,4,6,8])
y=np.array([1,4,16,36,64])
#plot function to be used to mark points
#mec- marker edge color the boundary(perimeter) , mfc=marker face color means all the area inside the boundary, ms: marker size
plt.plot(x,y,'o',marker='o',color='blue',ms=30,mfc='g',mec='r')
plt.show()
```

```python
#dotted lines
x=np.array([1,2,3])
y=np.array([2,3,4])
#colour is ls color,ls- linestyle  c -  line color  lw - line width
plt.plot(x,y,ls="--",marker="d",color='b',mfc="g",mec="r",ms=10,lw=3)
```

```python
#naming x and y axis
x=[1,2,3,4]
y=[30,43,23,15]
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.plot(x,y,marker='h', color='r',mfc='g',mec='c',ms=8,lw=3,ls=":")
```

```python
# Shorthand notation
x = [1,2,3,4,5]
y = [20,50,28,35,50]
plt.xlabel('Days')
plt.ylabel("Temperature in Degree")
# fmt = '[color][marker][line]'
plt.plot(x,y, 'b^--', label='2X')
```

```python
# X, Y axis Tickmarks (scale of your graphon x and y axis resp)

plt.xticks([10,20,30,40,50,60])
plt.yticks([10,20,30,40,50,60,70,80,90])
```

```python
days=[1,2,3,4,5,6,7]
temp1=[56,46,58,43,37,45,50]
temp2=[50,40,50,38,33,40,45]
avg=[53,43,54,40.5,35,42.5,47.5]
# figsize=(width,height) how big our image should look,dpi specifies pixels per inch
# #always guve figure size before plotting
plt.figure(figsize=(10,8),dpi=75) 
plt.plot(days,temp1)
plt.plot(days,temp2)
plt.plot(days,avg)
plt.xlabel("Days")
plt.ylabel("Temp")
plt.title("Weather")
# Save figure (dpi 300 is good when saving so graph has high resolution)
plt.savefig("Temp for a week.png",dpi=200)
plt.show()
```

```python
#to have a legend to show what each line refers to
days=[1,2,3,4,5,6,7]
temp1=[56,46,58,43,37,45,50]
temp2=[50,40,50,38,33,40,45]
avg=[53,43,54,40.5,35,42.5,47.5]
plt.title("Weather")

plt.plot(days,temp1,label="maxtemp")
plt.plot(days,temp2,label="mintemp")
plt.plot(days,avg,label="avgtemp")
plt.legend(loc='lower left')
#upper right, upper left, lower right, lower left, 
```

### Font Properties for Title and Labels

```python
day=[1,2,3,4,5,6,7]
temp=[50,51,52,48,47,49,46]
plt.plot(day,temp)
#fondict, labelpad, keyword args
plt.xlabel('Days',{'size':20,'color':'g'},labelpad=-140, color='red')
plt.show()
#note: the color mentioned outside i.e., red is taken into consideration 
```

```python
day=[1,2,3,4,5,6,7]
temp=[50,51,52,48,47,49,46]
plt.plot(day,temp)
plt.title("Weather", loc="right")
#fondict, labelpad, keyword args
plt.xlabel('Days',{'size':20,'color':'g'},labelpad=-140, color='red')
plt.show()
```

```python
#add grid lines
day=[1,2,3,4,5,6,7]
temp=[50,51,52,48,47,49,46]
plt.plot(day,temp)
plt.grid(color="g", lw="1",ls="--")
plt.show()
```

### Subplot

```python
#subplot (#no of row, #no of cols, #plot no)
##Subplot(1,3,1) - 1 row, 3 cols and first plot
#to add title to the fig we say supetitle("title")
days=[1,2,3,4,5,6,7]
temp1=[56,46,58,45,40,45,56]


plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.plot(days,temp1)
plt.title("Plot1")

temp2=[50,45,57,43,37,43,50]
plt.subplot(1,2,2)
plt.plot(days,temp2)
plt.title("Plot2")

plt.suptitle("Weather")

plt.show()
```

### Pie Charts

```python
x1=[10,5,5,45,35]
labels1=['color1','color2','color3','color4','color5']
plt.pie(x1,labels=labels1,startangle=90)
plt.show()
```

```python
#for projection and different colors
x1=[10,5,5,45,35]
labels1=['color1','color2','color3','color4','color5']
exp_value=[0, 0.3, 0, 0.2,0]
mycolor=['magenta', 'cyan','yellow','red','pink']
plt.pie(x1,labels=labels1,explode=exp_value,shadow=True,colors=mycolor)
plt.legend(title="color",loc="lower left")
plt.show()
```
