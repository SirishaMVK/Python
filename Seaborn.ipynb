## Plotting with the relplot function
It is a relational plot

```python
tips=sns.load_dataset("tips")
tips
```

```python
#Kind=scatter or lineplot, default is scatter
sns.relplot(data=tips,x="total_bill",y="tip",kind="line")
```

```python
sns.relplot(data=tips,x="total_bill",y="tip",hue="day",palette='viridis')
```

```python
#Aspect is the width/ht of facet
#height is ht of facet
sns.relplot(data=tips,x="total_bill",y="tip",hue="sex",aspect=2/1,height=5)
```

```python
#col_wrap - int; optional - “Wrap” the column variable at this width, so that the column facets span multiple rows.
#for example days, we have 4 days here thur,fri,sat,sun so it is across 4 graphs with how many in a row
#for example in time, we have 2 values i.e., lunch, dinner
sns.relplot(data=tips,x="total_bill",y="tip",hue="sex",col="time",col_wrap=2)
sns.relplot(data=tips,x="total_bill",y="tip",hue="sex",col="day",col_wrap=4)
```

```python
#hue_order is the order in which we want the data to be displayed
titanic = sns.load_dataset("titanic")
hue_order = ['Third', 'Second', 'First']
#kind = “point”, “bar”, “strip”, “swarm”, “box”, “violin”, or “boxen”.
sns.catplot(x="sex", y="survived", hue="class", data=titanic, hue_order=hue_order, kind="bar")
```

## Histogram -data distribution by forming bins

```python
df=sns.load_dataset("iris")
df.head(6)
```

### Kernel Density Estimate - visualizing the Probability Density of the continuous or non-parametric data variables

```python
#histogram is the distance plot/ displot as there's a short distance between the bars of the graph
a=sns.displot(df["petal_length"], kde=False)
print(a)
#to make the wave visible in the graph we give kde=true
b=sns.displot(df['petal_length'],kde="True")
print(b)
```

## Bar plot

```python
df1=sns.load_dataset("titanic")
#ci Confidence intervals how much higher or lower the percent could be it is like +or - of cetain value.
#run a , b to find the different values for "ci"
#a=sns.barplot(x="embark_town", y="age", data=df1, palette="BuPu",ci="sd")
hue_order=["Cherboug","Southampton","Queenstown"]
b=sns.barplot(x="embark_town", y="age", data=df1, palette="BuPu",hue_order=hue_order,ci=None)
```

```python
#capsize is caption size
#mentioned the legend as the aligment was improper
sns.barplot(x='sex',y='survived',hue='class',data=titanic, palette="YlOrBr",
            order=['male','female'],
           saturation=8,
           errcolor='magenta',
           errwidth='3',
           ci="sd")
plt.legend()
plt.show()
```

## Count plot

## histogram across a categorical variable

```python
#Cannot pass values for both `x` and `y`, as it a count of value
#always give figure size before plotting
plt.figure(figsize=(15,5),dpi=75)
sns.countplot(x='class', hue='who',data=df1,palette="magma")
plt.title("Survivors",color="Purple",size=25)
```

## Regplot

## Box Plot

## depicting groups of numerical data through the quartiles
Minimum -bottom black horizontal line of the box plot

First Quartile or 25% -first black horizontal line of the rectangle shape of the box plot

Median (Second Quartile) or 50% - second black horizontal line of the rectangle shape of the box plot

Third Quartile or 75% - third black horizontal line of rectangle shape

Maximum - top black horizontal line of the rectangle shape

outlier data - small diamond shape of the box plot

```python
#the black diamond spots are the out-layered data if this data is huge then we need to work on it else we can ignore them
sns.boxplot(x="day",y='total_bill',data=tips)
```

## Heatmap
two-dimensional graphical representation of data where the individual values that are contained in a matrix are represented as colours.

```python
f1=sns.load_dataset("flights")
f2=f1.pivot("month","year","passengers")
f2
```

```python
plt.figure(figsize=(15,5))
sns.heatmap(f2,linewidth=2,cmap='YlGnBu')
```

```python
flights=sns.load_dataset("flights")
corr=flights.corr()
print(corr)
sns.heatmap(corr,annot=True,linewidths=.5,cmap="YlGnBu")
```

## Cluster map
plots a matrix dataset as a hierarchically-clustered heatmap

```python
f1=sns.load_dataset("flights")
f2=f1.pivot("month","year","passengers")
sns.clustermap(f2,linewidth=2,cmap='YlGnBu')
```

## Facetgrid

```python
tips = sns.load_dataset("tips")
g = sns.FacetGrid(tips, col="time")
g.map(sns.scatterplot, "total_bill", "tip")
```

## Pair Plot
creates a grid of Axes of numeric variable in data

```python
df = sns.load_dataset('iris')
sns.set_style("ticks")
sns.pairplot(df,hue = 'species',diag_kind = "kde",kind = "scatter",palette = "husl")
plt.show()
```
